## Selecting data

We can select particular rows using standard Python array slicing notation:

```python
data[100:105]
```

To select both rows and columns using integers, the iloc attribute can be used with the format `.iloc[rows, columns]`

```python
data.iloc[100:105,0:5]
```

We can also select both columns and rows based on labels or conditions using `.loc[row_indexer,column_indexer]`

```python
data.loc[data['iso_code'] == 'QAT',['date','new_cases']]
```

There is also the possibility of using column label as an attribute to select the column:

```python
data.new_cases
```

Subnational locations (England, Northern Ireland, Scotland, Wales, Northern Cyprus…) and international aggregates (World, continents, European Union…) can be identified by their iso_code that starts with OWID_. We want to drop these entities from our data:

```python
data = data[~data['iso_code'].str.startswith('OWID_')]
```

We can also change the column `date` from string to datetime data type to make sure pandas recognizes its content as dates:

```python
data.loc[:,'date'] = pd.to_datetime(data.loc[:,'date'])
```

Let's say we don't want to carry around the whole dataset so we want to keep only some specific columns. To select columns, we can also pass a list containing the names of the desired columns:

```python
# Select desired columns
data = data[['iso_code','location','continent','date','total_cases','new_cases','total_cases_per_million','total_deaths','new_deaths','total_deaths_per_million','total_tests','total_tests_per_thousand','people_vaccinated','people_vaccinated_per_hundred','people_fully_vaccinated','people_fully_vaccinated_per_hundred','population','gdp_per_capita']]
```

We can create a new column using data from other columns easily:

```python
# create a new column
data['case_fatality_rate'] = data['total_deaths_per_million'] / data['total_cases_per_million']
```

```python
data
```

Let's say now we don't need the column `continent`. We can drop it from the dataframe:

```python
# drop a column
data = data.drop(['continent'], axis=1)
```

```python
# descriptive statistics
data.describe()
```

```python
# select the data with a date range using a Boolean condition
data[(data['date'] >= '2021-10-01')]
```

```python
# we can also combine conditions
data[(data['date'] >= '2021-10-01') & (data['people_fully_vaccinated_per_hundred'] >= 70)]
```

## Working with indices

Defining an index or a multi-index for a dataframe can be very useful in accessing the data efficiently. We can index the dataframe using dates and countries:

```python
data = data.set_index(['iso_code','date'],verify_integrity=True) # verify that the multindex is unique
```

```python
data
```

```python
data.loc['QAT',:]['total_cases']
```

```python
data.loc['QAT','2020-03-02']['total_cases']
```

In order to use slicing we need to sort the data based on the two indices:

```python
data.sort_index(level=['iso_code','date'], ascending=[1, 1], inplace=True)
```

Now we can do all sort of fancy slicing using `.IndexSlice`

```python
idx = pd.IndexSlice
data.loc[idx[['QAT','USA'],'2021-01-01':'2021-03-31'],:]
```

## Grouping and aggregating data

```python
# Select last available observation for each country
groupby_cntry = data.groupby('iso_code')
groupby_cntry
```

```python
groupby_cntry.groups.keys()
```

```python
# we can store it as list for later use
cntry_codes = list(groupby_cntry.groups.keys())
```

We can aggregate the data by group for all variables using some function. We can use `mean()` for example:

```python
groupby_cntry.mean()
```

We can also apply the aggregation to one or more specific columns:

```python
groupby_cntry['case_fatality_rate'].mean()
```

We can apply many aggregations at the same time using the `.agg` method:

```python
groupby_cntry['case_fatality_rate'].agg(['mean', 'median', 'std', 'count'])
```

Because the dataframe contains data reported by each government in a given, we would like to select for each government the latest available data:

```python
data_last = groupby_cntry.last()
```

We can now compute the share of people fully vaccinated around the world using the most recent data:

```python
data_last['people_fully_vaccinated'].sum() / data_last['population'].sum() * 100
```

## Merging data

We would like to add for each country its income group based on World Bank classification. This list is available in different file:

```python
# read data
temp = pd.read_csv('income_groups.csv', sep=',',header=0)
temp.head()
```

```python
temp = temp[~temp['Code'].str.startswith('OWID_')]  # make sure there are no international entities
temp = temp.reset_index(drop=True)
temp = temp.set_index(['Code'],verify_integrity=True)
```

We want to merge the our main dataframe with the column `Income group` of the `temp` dataframe. We are going to merge based on countries. So better check if the number of countries is same in both dataframes:

```python
len(temp.index)
```

```python
len(cntry_codes)
```

Because we have more countries in the main dataframe we are going to use a left merge:

```python
data_last = pd.merge(left=data_last, right=temp['Income group'], how='left', left_on=['iso_code'], right_index=True)  # choose inner for intersection
```

```python
data_last
```

Now we can compute and plot the average rate of vaccination for each group:

```python
share_fvax = data_last.groupby(['Income group'])['people_fully_vaccinated'].sum() / data_last.groupby(['Income group'])['population'].sum() * 100
share_fvax.plot(kind='bar', title='Share of population fully vaccinated', ylabel='%', xlabel='Income group', figsize=(6, 5))
```

We can also compute how many countries vaccinated more than 70% of their population:

```python
data_last[data_last['people_fully_vaccinated_per_hundred'] >= 70]['location'].count()
```

## Reshaping data

```python
# read data
temp = pd.read_csv('vaccinations-by-manufacturer.csv', sep=',',header=0)
temp.head()
```

We would like to create columns for each vaccine type using `.pivot_table`:

```python
data_vax = pd.pivot_table(temp, values='total_vaccinations', index=['location','date'], columns='vaccine')
```

```python
data_vax = data_vax.sort_index(level=['location','date'], ascending=[1, 1])
data_vax
```

```python
vax_dist = data_vax.groupby('location').last().sum()

vax_dist.plot.pie(figsize=(9, 6),autopct='%1.1f%%',labels=vax_dist.index,ylabel='Vaccine share by manifacturer')
```
